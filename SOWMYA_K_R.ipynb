4 .Fill in all the missing values present in all the columns in the dataset

```python
#checking null values
data.isna().sum()
```

```python
#plotting the graph
freqgraph=data.select_dtypes(include='float64')
freqgraph.hist(figsize=(20,15))
```

```python
#filling the missing values
data['Age']=data['Age'].fillna(data['Age'].mean())
data['Cabin']=data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
```

5 .. Check and handle outliers in at least 3 columns in the dataset

```python
num_columns=['Survived','Pclass','Age','SibSp','Parch','Fare']
```

```python
#checking outliers
for i in num_columns:
  plt.figure()
  plt.boxplot(data[i])
  plt.title(i);
```

```python
data.columns
```

Outliers - 'Age'

```python
Q1=np.percentile(data['Age'],25,method='midpoint')
Q2=np.percentile(data['Age'],50,method='midpoint')
Q3=np.percentile(data['Age'],75,method='midpoint')

print("Q1 is: ",Q1)
print("Q2 is: ",Q2)
print("Q3 is: ",Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR

print("lower limit:",low_lim)
print("upper limit:",up_lim)
```

```python
outlier =[]
for x in data['Age']:
  if((x<low_lim)or(x>up_lim)):
    outlier.append(x)

print(outlier)
```

```python
data['Age']=data['Age'].clip(lower=low_lim,upper=up_lim)
```

```python
outlier =[]
for x in data['Age']:
  if((x<low_lim)or(x>up_lim)):
    outlier.append(x)

print(outlier)
```

```python
plt.boxplot(data['Age'])
plt.show();
```
